```
from pyspark.sql.functions import udf, collect_list
from pyspark.sql.types import FloatType
import numpy as np
```

```
# Criando dados aleatórios
data = [
  [1, 12345, -1],
  [2, 12345, -2],
  [3, 12345, -3],
  [1, 54321, 3],
  [2, 54321, 1],
  [3, 54321, 2],
  [3, 54321, 4],
]

columns = ['month', 'customer_id', 'point']

dataframe = spark.createDataFrame(data, columns)
display(dataframe)
```

```
# Função para encontrar mediana
def get_mediana(values_list):
    try:
      # Usando função do numpy para encontrar mediana na lista
      median = np.median(values_list)
      # Arredondando valores
      return round(float(median),2)
    except Exception:
      return None

# Função para usar em colunas do df
get_mediana_udf = udf(get_mediana,FloatType())
```

```
dataframe_list = dataframe.groupBy("customer_id").agg(collect_list("point").alias("points"))
```

```
dataframe_median = dataframe_list.withColumn("median",get_mediana_udf("points"))
display(dataframe_median)
```
