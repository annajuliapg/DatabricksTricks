```
from multiprocessing.pool import ThreadPool
import time
```

```
def funcao(numero):
  
  # Para testar o paralelismo
  time.sleep(1)
  
  return numero
```

```
# Quantidade de Thread
pool = ThreadPool(5)

# Parametros da função
parameters = [x for x in range(10)]
```

```
# Rode as tarefas
results = pool.map(lambda numero: funcao(numero), parameters)
```

```
# Sem Thread
# Observe o tempo de execução comparando com a célula anterior
results_normal = []
for x in parameters:
  results_normal.append(funcao(x))
```
